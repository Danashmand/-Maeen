```python
import VirtualAssistant as va
import time as t
def testStories(level): 
    # Test the stories
    
    tt = t.time()
    story = va.getStory(level)
    print("Time taken to get story: ", t.time()-tt)
    # clean white spaces
    # story = story.strip()
    numWords = len(story.split())
    print("number of words: ", len(story.split()))
    print("number of characters: ", len(story.strip()))
    return numWords 

def mainReading():
    lst = []
    
    for i in range(1, 100,10):
        print("="*30)
        print("Level: ", i)
        su = 0
        for _ in range(5):
            a = testStories(i)
            su += a
        lst.append((i,su/5))
    
    # scatter plot   
    # with lines of the avg vs level
    import matplotlib.pyplot as plt
    import numpy as np
    lst = np.array(lst)
    x = lst[:,0]
    y = lst[:,1]
    plt.scatter(x,y)
    plt.xlabel("Level")
    plt.ylabel("Number of words")
    plt.title("Number of words in stories at each level")  
    plt.show()
    #  
            
    
mainReading()
```
